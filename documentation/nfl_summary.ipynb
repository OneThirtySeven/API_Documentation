<h3 style="margin-bottom: 0px; font-weight: bold;">Game Summaries</h3>

- Returns list of the 32 NFL teams.

***
GET: **https://nfl.onethirtyseven.io/summary**<br>

- Requires API Key<br>
    - To get your OneThirtySeven API Key follow the [instructions](GET_API_KEY.md):
- No optional parameters
***

```python
import requests
import pandas as pd
import json

# Replace with your personal API_KEY
API_KEY = '3dc348e87a2988b1b90f5137313d436f57e258bceb9b4e7f643a6fb8fa96b77e'

# Make get request
res = requests.get('https://nfl.onethirtyseven.io/summary', headers={'api_key': API_KEY})

# Extract payload: Decode binary -> Convert the JSON str to dict() -> Convert dict to DataFrame
Game_Summaries = pd.DataFrame.from_dict(json.loads(res.content.decode('utf-8')))

# Display Regular Season Eagles Games
pd.set_option('display.max_columns', None)
Game_Summaries[(Game_Summaries['Season_Type'] == 'Regular-Season') & (Game_Summaries['Team'] == 'PHI')].head(17)
```
