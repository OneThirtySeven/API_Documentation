<h3 style="margin-bottom: 0px; font-weight: bold;">Player Stats</h3>

- Returns Player Stats on a per game basis

***
GET: **https://nfl.onethirtyseven.io/player_box**<br>

- Requires API Key<br>
    - To get your OneThirtySeven API Key follow the [instructions](GET_API_KEY.md):
- Optional Parameters
    - GameID
***

```python
%%capture --no-display
import pandas as pd
pd.set_option('display.max_columns', None)
```

```python
import requests
import pandas as pd
import json

# Replace with your personal API_KEY
API_KEY = '3dc348e87a2988b1b90f5137313d436f57e258bceb9b4e7f643a6fb8fa96b77e'

# Make get request
res = requests.get('https://nfl.onethirtyseven.io/player_box?GameId=803098376', headers={'api_key': API_KEY})

# Extract payload: Decode binary -> Convert the JSON str to dict() -> Convert dict to DataFrame
Player_Stats = pd.DataFrame.from_dict(json.loads(res.content.decode('utf-8')))

# Display Top 5 by Passing_Yards
Player_Stats.sort_values('Pass_Yards', ascending=False).head(5)
```
